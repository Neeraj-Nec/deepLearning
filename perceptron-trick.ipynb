```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_classes=2, n_informative=2, n_redundant=0, class_sep=10, hypercube=False, n_clusters_per_class=1, random_state=41)
```

```python
print(y)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolors='k')
```

```python
def step(x):
    return 1 if x >= 0 else 0
```

```python
def perceptron(X, y):
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    for i in range(10000):
        j = np.random.randint(1,100)
        y_hat = step(np.dot(X[j], weights))
        weights = weights + (y[j]-y_hat)*X[j]
    return weights[0], weights[1:]
    
```

```python
intercept_, coef_ = perceptron(X,y)
```

```python
intercept_
```

```python
coef_
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input+b
```

```python
plt.figure(figsize = (10,6))
plt.plot(x_input, y_input, color = 'red', linewidth =3)
plt.scatter(X[:,0], X[:,1], c=y, cmap = 'winter',s=100)
plt.ylim(-3,2)
```
