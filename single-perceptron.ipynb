```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/kaggle/input/placement/Placement.csv")

print(df.head())
```

```python
df.shape
```

```python
sns.scatterplot(x=df['CGPA'], y=df['IQ'], hue=df['Placement'])
```

```python
x = df.iloc[:, 1:3]
print(x)
print(x.values)
```

```python
y = df.iloc[:,-1]
print(y)
print(y.values)
```

```python
from sklearn.linear_model import Perceptron

# Create the model instance
p = Perceptron()

# Fit the model with training data (x = features, y = labels)
p.fit(x, y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.values, y.values, clf = p, legend =2)
```
